```python
import pandas as pd

# Convert the 'date_column' to datetime if it's not already
df['Date'] = pd.to_datetime(df['Date'])

# Create a filter to keep rows with dates from 2012 or later
filter_condition = df['Date'].dt.year >= 2012

# Apply the filter to keep only the desired rows
df = df[filter_condition].reset_index()
```

```python
# Dates to replace
aa = ['07-10-2562', '03-08-2015']

# Replace date values in 'VOLUME' with 0
df['VOLUME'] = df['VOLUME'].replace(aa, 0).astype('float64')


# Step 1: Calculate the total volume for each company
total_volume_by_company = df.groupby('TRADING CODE')['VOLUME'].sum().reset_index()

# Step 2: Sort the DataFrame based on the total volume
sorted_companies = total_volume_by_company.sort_values(by='VOLUME', ascending=False)

# Step 3: Select the top 100 companies
top_100_companies = sorted_companies.head(100)

# Step 4: Filter the original DataFrame to keep data only for these top 100 companies
df = df[df['TRADING CODE'].isin(top_100_companies['TRADING CODE'])]
df = df.drop(columns=['index'])
df
```

```python
# Convert 'OPEN' and 'CLOSE' columns to numeric
df['OPEN'] = pd.to_numeric(df['OPEN'], errors='coerce')
df['CLOSE'] = pd.to_numeric(df['CLOSE'], errors='coerce')

# Remove rows with missing or non-numeric values
df.dropna(subset=['OPEN', 'CLOSE'], inplace=True)

# Continue with your analysis or modeling steps
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA

# Sort the DataFrame by 'Date'
df.sort_values(by='Date', inplace=True)

# Step 2: Create lag features
df['lag1'] = df['OPEN'].shift(1)

# Step 3: Train the ARIMA model
model = ARIMA(df['CLOSE'], order=(5, 1, 0))
model_fit = model.fit()

# Function to predict closing price given a trading code, date, and opening price
def predict_closing_price(trading_code, date, opening_price):
    # Create a DataFrame for prediction input
    prediction_data = pd.DataFrame({'Date': [date], 'OPEN': [opening_price]})
    prediction_data['Date'] = pd.to_datetime(prediction_data['Date'])

    # Ensure that 'OPEN' is a numeric value
    prediction_data['OPEN'] = pd.to_numeric(prediction_data['OPEN'], errors='coerce')

    # Use the ARIMA model to forecast the closing price
    forecast = model_fit.forecast(steps=1, exog=prediction_data['OPEN'].values.reshape(-1, 1))
    return forecast

# Example: Predict closing price for a specific trading code, date, and opening price
trading_code_to_predict = '00DSEGEN'
date_to_predict = '2012-06-01'
opening_price_to_predict = 5500.00
predicted_closing_price = predict_closing_price(trading_code_to_predict, date_to_predict, opening_price_to_predict)
print(f"Predicted Closing Price for {trading_code_to_predict} on {date_to_predict}: {predicted_closing_price}")
```

```python
actual= 5467.0

# Calculate percentage error
percentage_error = ((actual - predicted_closing_price) / actual) * 100

# Calculate mean percentage error
mean_percentage_error = percentage_error.mean()

# Print the mean percentage error
print(f"Mean Percentage Error: {mean_percentage_error:.2f}%")
```

```python
import pandas as pd
import numpy as np
from fbprophet import Prophet



# Extract features (date, trading code, and opening price)
X = df[['Date', 'TRADING CODE', 'OPEN']]
# Target variable (closing price)
y = df['CLOSE']

# Create lag features for opening price
X['lag1'] = X['OPEN'].shift(1)

# Rename columns for Prophet
X.rename(columns={'Date': 'ds', 'CLOSE': 'y'}, inplace=True)

# Encode the 'TRADING CODE' feature as categorical
X['TRADING CODE'] = X['TRADING CODE'].astype('category')

# Remove missing values
X.dropna(inplace=True)
y = y[X.index]

# Create and train the Prophet model
model = Prophet()
model.add_regressor('lag1')
model.fit(X)


# Function to predict closing price given a date, trading code, and opening price
def predict_closing_price(date, trading_code, opening_price):
    # Create a DataFrame for prediction input
    prediction_data = pd.DataFrame({'ds': [date], 'TRADING CODE': [trading_code], 'OPEN': [opening_price]})
    prediction_data['TRADING CODE'] = prediction_data['TRADING CODE'].astype('category')
    prediction_data['lag1'] = X['lag1'].iloc[-1]

    # Make future DataFrame for prediction
    future = model.make_future_dataframe(periods=1)
    future['TRADING CODE'] = trading_code
    future['lag1'] = prediction_data['lag1'].values[0]

    # Forecast with Prophet
    forecast = model.predict(future)

    # Get the predicted closing price
    predicted_closing_price = forecast['yhat'].iloc[-1]
    return predicted_closing_price

# Example: Predict closing price for a specific date, trading code, and opening price
date_to_predict = '2012-06-01'
trading_code_to_predict = '00DSEGEN'
opening_price_to_predict = 5500.00
predicted_closing_price = predict_closing_price(date_to_predict, trading_code_to_predict, opening_price_to_predict)
print(f"Predicted Closing Price for {date_to_predict} ({trading_code_to_predict}): {predicted_closing_price}")
```
